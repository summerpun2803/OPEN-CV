```python
img_grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
imgplot = plt.imshow(img_grey,cmap='gray')
```

```python
def gaussian(sig,size):
    size=int(size)//2
    x,y=np.mgrid[-size:size+1 ,-size:size+1]
    z = 1/(2* np.pi* (sig**2))
    g = np.exp(-((x**2 +y**2)/(2 * (sig**2))))*z
    return g
    
```

```python
image_blurred = ndimage.convolve(img_grey,gaussian(12, 5))
imgplot = plt.imshow(image_blurred,cmap='gray')
```

```python
def sobel(img):
    ky = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
    kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
    
    Ix = ndimage.convolve(img, kx)
    Iy = ndimage.convolve(img, ky)

    G = np.hypot(Ix, Iy)
    theta = np.arctan2(Iy, Ix)
    
    return G , theta
```

```python
mag , angle=sobel(image_blurred)
imgplot = plt.imshow(mag,cmap='gray')
```

```python
def suppresion(img , angle):
    M,N = img.shape
    theta = angle*180/np.pi
    theta[theta<0] = theta[theta<0] + 180
    
    matrix = np.zeros((M,N))
    a = 255
    b = 255
    for i in range (1,M-1):
        for j in range (1,N-1):
            
            if (0<= theta[i,j] < 22.5) or (157.5 < theta[i,j]<=180):
                a = img[i,j+1]
                b = img[i,j-1]
            elif (22.5<= theta[i,j]<67.5):
                b = img[i+1,j-1]
                a = img[i-1,j+1]
            elif (67.5<=theta[i,j]<112.5):
                a = img[i-1,j]
                b = img[i+1,j]
            elif (112.5<=theta[i,j]<=157.5):
                a = img[i-1,j-1]
                b = img[i+1,j+1]
            
            if (img[i,j]>= a and img[i,j]>=b):
                matrix[i,j] = img[i,j]
            

    return matrix
```

```python
Z = suppresion(mag,angle)
```

```python
def thresholding(img):
    max=img.max()*0.9
    min=max*0.3
    
    M,N = img.shape
    matrix = np.zeros((M,N))
    
    strong = 255
    weak = 100
            
    for i in range(0,M-1):
        for j in range(0,N-1):
            if img[i,j]>=max:
                matrix[i,j] = strong
            elif img[i,j]<max and img[i,j]>=min:
                matrix[i,j] = weak
    return matrix
```

```python
mat = thresholding(Z)
plt.hist(mat.ravel(),256,[0,360])
plt.show()
```

```python
imgplot = plt.imshow(mat,cmap='gray')
```

```python
def hysterisis(img):
    weak = 100
    strong = 255
    
    M,N=img.shape
    for i in range (M):
        for j in range (N):
            
            if img[i,j]==weak:
                
                if (img[i+1,j]==strong or img[i-1,j] == strong or img[i,j+1] == strong or img[i,j-1]==strong or img[i+1,j+1] == strong or img[i+1,j-1] == strong or img[i-1,j-1]==strong or img[i-1,j+1]== strong):
                    img[i,j] = strong
                
                else:
                    img[i,j] = 0
    return img
    
```

```python
mat1 = hysterisis(mat)
plt.hist(mat1.ravel(),256,[0,360])
plt.show()
```
